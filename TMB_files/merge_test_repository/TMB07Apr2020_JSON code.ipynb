# JSON Reader


```python
import json
import matplotlib.pyplot as plt
from datetime import datetime

with open('n4023f_stamprun.json') as f:
  data = json.load(f)
print(len(data))
```

```python
## Separate data into usable lists

lat_List = []
course_List = []
long_List = []
alt_List = []
speed_List = []
time_List = []

for i in range(0,len(data)):
    lat_List.append(data[i]["Latitude"])
    course_List.append(data[i]["Course"])
    long_List.append(data[i]["Longitude"])
    alt_List.append(data[i]["Altitude"])
    speed_List.append(data[i]["Speed"])
    time_List.append(datetime.strptime(data[i]["Timestamp"], "%Y-%m-%dT%H:%M:%S.%fZ"))

print(len(lat_List))
print(len(course_List))
print(len(long_List))
print(len(alt_List))
print(len(speed_List))
print(len(time_List))
print(time_List[0])
```

```python
plt.plot(time_List,lat_List)
plt.xlabel('Timestamp')
plt.ylabel('Latitude (°)')
plt.title('Latitude over Time')
plt.show()
```

```python
plt.plot(time_List,long_List)
plt.xlabel('Timestamp')
plt.ylabel('Longitude (°)')
plt.title('Longitude over Time')
plt.show()
```
