```python
import pandas as pd
import matplotlib.pyplot as plt

dataList = []
fileData = pd.read_csv("flight_maps-master/sample_flight_sim.csv")
print(fileData)
```

```python
time_List = fileData.Time
long_List = fileData.Longitude
lat_List = fileData.Latitude
alt_List = fileData.Altitude
roll_List = fileData.Roll
pitch_List = fileData.Pitch
heading_List = fileData.Heading
```

```python
plt.plot(time_List,lat_List)
plt.xlabel('Timestamp')
plt.ylabel('Latitude (°)')
plt.title('Latitude over Time')
plt.show()
```

```python
plt.plot(time_List,long_List)
plt.xlabel('Timestamp')
plt.ylabel('Longitude (°)')
plt.title('Longitude over Time')
plt.show()
```

```python
plt.plot(time_List,alt_List)
plt.xlabel('Timestamp')
plt.ylabel('Altitude')  #Units not mentioned
plt.title('Altitude over Time')
plt.show()
```

```python
plt.plot(time_List,roll_List)
plt.xlabel('Timestamp')
plt.ylabel('Roll (°)')
plt.title('Roll over Time')
plt.show()
```

```python
plt.plot(time_List,pitch_List)
plt.xlabel('Timestamp')
plt.ylabel('Pitch (°)')
plt.title('Pitch over Time')
plt.show()
```

```python
plt.plot(time_List,heading_List)
plt.xlabel('Timestamp')
plt.ylabel('Heading (°)')
plt.title('Heading over Time')
plt.show()
```
